```python
#1
def sum_of_evens(nums):
    total = 0
    for n in nums:
        if n % 2 == 0:
            total += n
    return total

# Example
print(sum_of_evens([1, 2, 3, 4, 6]))  # Output: 12
```

```python
#4
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example
print(is_prime(17))  # Output: True
print(is_prime(20))  # Output: False
```

```python
#5
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

# Example
for num in Fibonacci(7):
    print(num, end=" ")
```

```python
#6
def powers_of_two(exp):
    for i in range(exp + 1):
        yield 2 ** i

# Example
for val in powers_of_two(5):
    print(val, end=" ")
```

```python
#7
def read_file(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

# Example usage
# for line in read_file("sample.txt"):
#     print(line)
```

```python
#9
celsius = [0, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
# Output: [32.0, 68.0, 86.0, 104.0]
```

```python
#10
def remove_vowels(s):
    vowels = "aeiouAEIOU"
    return ''.join(filter(lambda ch: ch not in vowels, s))

# Example
print(remove_vowels("Hello World"))  # Output: Hll Wrld
```
